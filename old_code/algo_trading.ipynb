```python
my_portfolio = yf.Ticker("CCL") 
test_2_df = my_portfolio.history(period="1y", interval = "1d")
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
test_2_df.head()
```

```python
test_2_df['movement'] = test_2_df['Close'] - test_2_df['Open']
test_2_df['daily_return'] = test_2_df['Close'].dropna().pct_change()
test_2_df['Positive Return'] = np.where(test_2_df['daily_return'] > 0, 1.0, 0.0)
test_2_df.head()
```

```python
test_2_df['decision'] = np.nan
test_2_df['profit/loss'] = np.nan
test_2_df.head()
```

```python
previous_price = 0
```

```python
buy = []
sell = []

for index, row in test_2_df.iterrows():

    if previous_price == 0:
        test_2_df.loc[index, "decision"] = "buy"
        buy.append(row["Close"])
        test_2_df.loc[index, "profit/loss"] = 0
    elif row["Close"] < previous_price:
        test_2_df.loc[index, "decision"] = "buy"
        buy.append(row["Close"])
        test_2_df.loc[index, "profit/loss"] = 0
    elif row["Close"] > previous_price:
        test_2_df.loc[index, "decision"] = "sell"
        sell.append(row["Close"])
        test_2_df.loc[index, "profit/loss"] = sell[-1] - buy[-1]
    else:
        test_2_df.loc[index, "decision"] = "hold"

    previous_price = row["Close"]
```

```python
test_2_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
label_encoder = LabelEncoder()
label_encoder.fit(test_2_df['decision'])
list(label_encoder.classes_)
label_encoder.transform(test_2_df['decision'])
test_2_df['buy_sell_hold'] = label_encoder.transform(test_2_df['decision'])
test_2_df.sample(5)
```

```python
#Shifting independent variables 
x_var_list = ['Close', 'Volume', 'Open', 'High', 'Low', 'movement']
neural_df[x_var_list] = neural_df[x_var_list].shift(1).dropna()
neural_df.head()
```
